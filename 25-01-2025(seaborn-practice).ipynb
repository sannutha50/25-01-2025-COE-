```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Histogram
sns.histplot(data=tips,x='total_bill',bins=20,color="skyblue",kde=True)
plt.title("Distribution of Total Bills")
plt.show()
```

```python
#KDE plot
sns.kdeplot(data=tips, x="total_bill", fill=True, color="purple")
plt.title("KDE Plot of Total Bills")
plt.show()
```

```python
#Bar plot
sns.barplot(data=tips,x="day",y="total_bill",palette="colorblind")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
#Box plot
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box Plot of Total Bills by Day And Gender")
plt.show()
```

```python
#scatterplot
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",hue="day",marker="o",color="green")
plt.title("Line Plot of Party Size vs. Tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")

plt.show()
```

```python
#Heatmaps display data as a color-codd matrix
numeric_data = tips.select_dtypes(include=['number'])
correlation_matrix=numeric_data.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="hsv",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Seaborn on Train.csv dataset
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("train.csv")
df.head()
```

```python
sns.barplot(data=df,x='Pclass',y='Survived',errorbar=None,palette=['yellow','pink','cyan'])
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#is there a relationship b/w passenger age and fare
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',palette='summer',alpha=0.6)
plt.title("Relationship b/w Passenger Age and Fare")
plt.xlabel("Passenger Age")
plt.ylabel("Fare")
plt.show()
```

```python
#wt are the pairwise relationship b/w age,fare,and passenger class
subset=df[['Age','Fare','Pclass','Survived']]
sns.pairplot(subset,hue='Survived',palette='coolwarm',diag_kind='kde')
plt.suptitle('Pairwise Relationship',y=1.02)
plt.show()
```

```python
#wt is the correlation b/w different numerical features in dataset
numeric_data = df.select_dtypes(include=['number'])
correlation_matrix=numeric_data.corr()
sns.heatmap(correlation_matrix,annot=True,cmap="magma",fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
